```python
import pandas as pd
```

```python
mice_df = pd.read_csv("test_data/mice_data.csv")
worms_df = pd.read_csv("test_data/worms_data.csv")
```

```python
def get_max_avg_lifespan(df):
    grouped = df.groupby(["Compound/Formulation", "Species"])
    aggregation = grouped.aggregate({'Avg/Med Lifespan Change (%)': 'max'})
    res_df = aggregation.sort_values(by="Avg/Med Lifespan Change (%)", ascending=False)
    return res_df
```

```python
get_max_avg_lifespan(mice_df)
```

```python
get_max_avg_lifespan(worms_df)
```
